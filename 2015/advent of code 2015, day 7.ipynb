```python
def AND(a, b): return a & b
def OR(a, b): return a | b
def NOT(a): return 65535 - a
def LSHIFT(a, b): return a << b
def RSHIFT(a, b): return a >> b
```

```python
# advent of code 2015, day 7

wires = {}
processes = {}
dependencies = {}

steps = [[line.split()[:-2], line.split()[-1]] for line in data.split('\n')]

def helper1(wire): return int(wire) if wire.isdigit() else wires.get(wire, -1)

def helper2(instruction):
    if len(instruction) == 1:
        [wire_a] = instruction
        value_a = helper1(wire_a)
        return value_a if value_a >= 0 else (wire_a,)
    elif len(instruction) == 2:
        [func, wire_a] = instruction
        value_a = helper1(wire_a)
        return eval(f'{func}({value_a})') if value_a >= 0 else (wire_a,)
    else:
        [wire_a, func, wire_b] = instruction
        value_a, value_b = helper1(wire_a), helper1(wire_b)
        if value_a >= 0 <= value_b:
            return eval(f'{func}({value_a},{value_b})')
        else:
            values = []
            if value_a < 0: values += [wire_a]
            if value_b < 0: values += [wire_b]
            return tuple(values)

for i, (instruction, wire) in enumerate(steps[:]):
    value = helper2(instruction)
    if isinstance(value, tuple):
        processes[i] = value
        for item in value:
            dependencies[item] = dependencies.setdefault(item, []) + [i]
    else:
        wires[wire] = value

print(wires)
print(processes)
print(dependencies)

#for part in (1, 2):
#    answer = 
#    print(f'advent of code 2015, day 4, part {part}, answer: {answer}')
```
